```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('/content/highest.csv')  # Update the path if needed
print(df.head())  # Displays the first few rows of the dataset

# Scatter Plot
# Comparing Opening and Total Gross
plt.figure(figsize=(10, 6))
plt.scatter(df['Opening'], df['Total Gross'])
plt.title('Scatter Plot of Opening vs Total Gross')
plt.xlabel('Opening')
plt.ylabel('Total Gross')
plt.show()

# Bar Chart
# Average opening per Distributor
average_opening_per_distributor = df.groupby('Distributor')['Opening'].mean()
plt.figure(figsize=(10, 6))
average_opening_per_distributor.plot(kind='bar')
plt.title('Average Opening per Distributor')
plt.xlabel('Distributor')
plt.ylabel('Average Opening')
plt.show()

# Line Chart
# Assuming there's a 'Date' column formatted as DD-MM-YYYY, and you want to plot Total Gross over time
# Convert 'Date' to datetime format for proper plotting
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Total Gross'], marker='o')
plt.title('Total Gross Over Time')
plt.xlabel('Date')
plt.ylabel('Total Gross')
plt.xticks(rotation=45)
plt.show()

```
